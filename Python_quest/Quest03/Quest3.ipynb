n-gram만들기

```python
file_path = '/content/drive/MyDrive/data/06TheAvengers.txt'
with open(file_path, 'r', encoding='utf8') as f:
  lines = f.readlines()
```

```python
# word단위 딕셔너리 생성

import re
p = re.compile('[a-z]+')
dic = {}
for line in lines:
  line = line.lower()
  # line = line.split()
  line = p.findall(line)
  for i in line:
    if i in dic.keys():
      dic[i] += 1
    else:
      dic[i] = 1
```

```python
# 2-gram단위 딕셔너리 생성

import re
p = re.compile('[a-z]+')
dic = {}
for line in lines:
  line = line.lower()
  # line = line.split()
  line = p.findall(line)
  for i in range(len(line) -1):
    ngram = (line[i], line[i+1])
    if ngram in dic.keys():
      dic[ngram] += 1
    else:
      dic[ngram] = 1
```

```python
print(dic)
```

```python
# 리스트 컴프리헨션
def counter(dic):
  max2gram = [(k, v) for k, v in dic.items() if max(dic.values()) == v]
  return max2gram
```

```python
counter(dic)
```

```python
# for문
def counter(dic):
  for k, v in dic.items():
    if max(dic.values()) == v:
      max2gram = k, v
  return max2gram
```
