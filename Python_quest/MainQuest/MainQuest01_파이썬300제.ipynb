```python
import random

accounts_list = []
num_accounts = 0
bank_name = "SC은행"

# Q1. Account 클래스 생성
class Account:

    def __init__(self):
        owner = input("계좌를 개설할 예금주의 이름을 입력해주세요.\n 예: 홍길동 \n >")
        initial_balance = int(input("개설할 계좌의 초기 잔액을 입력해주세요.\n 숫자만 입력하세요. 예: 1000 \n >"))

        self.bank_name = bank_name
        self.owner = owner
        self.balance = initial_balance
        self.account_number = f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}" # 계좌번호를 랜덤하게 생성 (3자리-2자리-6자리 형태)
        self.balance_history = []  # 입금 내역을 저장할 리스트
        self.deposit_count = 0 # 입금 횟수

        # Q2. 계좌 객체 생성 시 클래스 변수 num_accounts를 1 증가시킴
        global num_accounts
        num_accounts += 1

        # Q8. 생성된 계좌를 리스트에 저장
        global accounts_list
        accounts_list.append(self)

    def __str__(self):
        return f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.format_balance()}원"

    # Q3. 생성된 계좌의 수를 출력
    def get_account_num():
        print("생성된 계좌 수 : ", num_accounts)

    # Q4. 입금 메서드
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.balance_history.append(f"[입금] {amount}원")
            print(f"{amount}원이 입금되었습니다.")
            self.deposit_count += 1
        if self.deposit_count % 5 == 0:
            self.check_interest()

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.balance_history.append(f"[츌금] {amount}원")
            print(f"{amount}원이 출금되었습니다.")
        else:
            print("출금 실패: 잔액 부족 또는 유효하지 않은 금액입니다.")

    # Q6. 정보 출력 메서드
    def display_info(self):
        print("=" * 5)
        print(f"은행이름: {self.bank_name}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.format_balance()}원")
        print("=" * 5)

    def format_balance(self):
        # 잔고를 세자리마다 쉼표로 구분하여 반환
        if self.balance < 1000:
            return self.balance  # 잔고가 3자리 이하일 경우 그대로 반환
        else:
            return f"{self.balance:,}"  # 잔고가 4자리 이상일 경우 밑줄로 구분하여 반환

    # Q7. 입금 횟수 매 5회마다 이자 지급
    def check_interest(self):
        interest = self.balance * 0.01
        self.balance += interest
        print(f"이자 지급: '{self.owner}'계좌에 이자 {interest}원 지급.")

    # Q9. 객체 순회하며 잔고가 100만원 이상인 고객 정보 출력
    def high_balance_check():
        for account in accounts_list:
            if account.balance >= 1000000:
                print("# 아래의 고객은 잔고가 100만원 이상입니다.")
                account.display_info()

    # Q10. 입출금 내역 출력
    def balance_history():
        print("=" * 3, "입출금 내역", "=" * 3)
        for account in accounts_list:
            history = account.balance_history
            for text in history:
                print("입출금 내역:", text)


new_account = Account()

new_account.deposit(5000)
new_account.withdraw(3000)
new_account.deposit(10000)
new_account.deposit(1000000)
new_account.deposit(200000)
new_account.deposit(10000)
new_account.deposit(10000)
new_account.withdraw(1000)


# 계좌 개수 출력
Account.get_account_num()

# 객체 순회하며 잔고가 100만원 이상인 고객 정보 출력
Account.high_balance_check()

# 입출금 내역 출력
Account.balance_history()
```

```python
import random
import csv

core_list = []
data_count = 0

# Aiffel 클래스 정의
class Aiffel:
    # 생성자 (이름, 입학연도, 기수, 과정명)
    def __init__(self, name, admission_year, order, course):
        self.name = name
        self.admission_year = admission_year
        self.order = order
        self.course = course
        self.quest_score = 0  # 퀘스트 점수 초기화
        self.total_score = 0  # 추가 점수 초기화
        self.penalty_score = 0  # 퀘스트 차감 점수 초기화
        self.student_id = self.generate_student_id()  # 학번 생성
        core_list.append({
            "이름": self.name,
            "학번": self.student_id,
            "과정 종류": self.course,
            "기수": self.order,
            "퀘스트 점수 총점": self.total_score,
            "퀘스트 입력 총점": self.quest_score,
            "퀘스트 penalty 총합": self.penalty_score

        })

    # 학번 생성 메서드
    def generate_student_id(self):
        year_last_two_digits = str(self.admission_year)[-2:]
        order_str = str(self.order).zfill(2)
        course_type = "1" if self.course == "리서치" else "2"
        random_number = str(random.randint(0, 99)).zfill(2)
        return year_last_two_digits + order_str + course_type + random_number

    # Q1. 사용자로부터 데이터 입력 받기
    def data_input(self):
        global data_count
        data_input = "김건: 0, 김경민: 0, 김소연: 0, 김영신: 0, 박근수: 0, 박영수: 0, 박영준: 0, 서원영: 0, 신지만: 0"
        # 입력된 문자열을 파싱하여 딕셔너리로 변환
        try:
            # 입력된 문자열을 쉼표(,)로 분할
            items = data_input.split(',')
            for item in items:
                # 각 아이템을 콜론(:)으로 분할하여 key와 value로 분리
                name, value = item.strip().split(':')
                Aiffel(name, 2023, "6기", "코어")
                data_count += 1
        except (ValueError, IndexError):
            print("올바른 형태로 데이터를 입력하세요.")

    # Q2. 학번의 개수를 출력하는 메서드
    def get_id_num(self):
        print("생성된 학번의 개수 :",data_count)

    # Q3. 퀘스트 점수 입력 메서드
    def score(self, name, points):
        if 1 <= points <= 3:
            # 반복문을 사용하여 리스트의 각 딕셔너리를 검사
            for member in core_list:
                # 딕셔너리에서 '이름' 키에 해당하는 value를 확인
                targer_name = member.get("이름")

                # 원하는 값을 찾았을 때 처리
                if name == targer_name:
                    # 딕셔너리의 '이름' 키에 새로운 이름 할당
                    member["퀘스트 입력 총점"] += points
                    member["퀘스트 점수 총점"] += points
                    print(f"{name}의 퀘스크 점수를 {points}점 추가하였습니다.")
                    print(member)
                    break
            else:
                # 반복문이 종료될 때까지 원하는 값을 찾지 못한 경우
                print(f"{name}의 이름을 가진 사람을 찾을 수 없습니다.")
        else:
            print("1~3 범위의 점수만 입력 가능합니다.")

    # Q4. 점수 차감 메서드
    def penalty(self, name, points):
        points = int(points)
        # 반복문을 사용하여 리스트의 각 딕셔너리를 검사
        for member in core_list:
            # 딕셔너리에서 '이름' 키에 해당하는 value를 확인
            targer_name = member.get("이름")

            # 원하는 값을 찾았을 때 처리
            if name == targer_name:
                # 딕셔너리의 '이름' 키에 새로운 이름 할당
                member["퀘스트 penalty 총합"] -= points
                member["퀘스트 점수 총점"] += points
                print(f"{name}의 퀘스크 점수를 {points}점 감점하였습니다.")
                print(member)
                break
        else:
            # 반복문이 종료될 때까지 원하는 값을 찾지 못한 경우
            print(f"{name}의 이름을 가진 사람을 찾을 수 없습니다.")


    # Q5. 정보 출력 메서드
    def display_info(self):
        for member in core_list:
            print(f"이름: {member['이름']}")
            print(f"학번: {member['학번']}")
            print(f"과정 종류: {member['과정 종류']}")
            print(f"퀘스트 점수 총점: {member['퀘스트 점수 총점']}")
            print(f"퀘스트 입력 총점: {member['퀘스트 입력 총점']}")
            print(f"퀘스트 penalty 총합: {member['퀘스트 penalty 총합']}")
            print("-" * 20)  # 구분선

        # CSV 파일로 출력
        with open('core_info.csv', 'w', newline='') as csvfile:
            fieldnames = ['이름', '학번', '과정 종류', '퀘스트 점수 총점', '퀘스트 입력 총점', '퀘스트 penalty 총합']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            for member in core_list:
                writer.writerow({
                    '이름': member['이름'],
                    '학번': member['학번'],
                    '과정 종류': member['과정 종류'],
                    '퀘스트 점수 총점': member['퀘스트 점수 총점'],
                    '퀘스트 입력 총점': member['퀘스트 입력 총점'],
                    '퀘스트 penalty 총합': member['퀘스트 penalty 총합']
                })

    # Q6. 중도포기자 삭제 메서드
    def remove_member(self, name):
        for member in core_list:
            if member['이름'] == name:
                core_list.remove(member)
                print(f"{name}님의 정보를 삭제하였습니다.")
                break
        else:
            print(f"{name}님을 찾을 수 없습니다.")



# Q7. Aiffel 클래스를 상속 Aiffel_Group 클래스 정의
class Aiffel_Group(Aiffel):
    def __init__(self, name, admission_year, order, course):
        super().__init__(name, admission_year, order, course)
        self.group_dict = {
            "상생": [],
            "신뢰": [],
            "열정": [],
            "이끔": [],
            "성장": [],
            "가치": [],
            "공유": [],
            "확산": [],
        }

    # 그룹에 멤버 추가
    def add_to_group(self, name):
        # 랜덤하게 그룹 선택
        group_names = list(self.group_dict.keys())
        selected_group = random.choice(group_names)
        # 멤버를 선택한 그룹에 추가
        self.group_dict[selected_group].append(name)
        print(f"{name}님이 {selected_group} 그룹에 추가되었습니다.")

    # 그룹핑 실행
    def group_members(self):
        members = [member["이름"] for member in core_list]
        while members:
            member = random.choice(members)
            self.add_to_group(member)
            members.remove(member)

    # 그룹 정보 출력
    def display_group_info(self):
        for group_name, group_members in self.group_dict.items():
            print(f"{group_name} 그룹: {', '.join(group_members)}")


# Q8. Aiffel 클래스를 상속하여 길드를 랜덤하게 매핑하는 Aiffel_Guild 클래스 정의
class Aiffel_Guild(Aiffel_Group):
    def __init__(self, name, admission_year, order, course):
        super().__init__(name, admission_year, order, course)
        # 길드 이름을 멤버 이름에 매핑하는 딕셔너리
        self.guild_members = {}
        # 길드별 점수를 저장하는 딕셔너리 초기화
        self.guild_score = {
            "힌튼": 0,
            "알트만": 0,
            "응": 0,
            "하사비스": 0,
        }
        # 길드 그룹 딕셔너리 초기화
        self.group_dict = {
            "힌튼": [],
            "알트만": [],
            "응": [],
            "하사비스": [],
        }

    # 그룹핑 실행
    def group_members(self):
        members = [member["이름"] for member in core_list]
        random.shuffle(members)  # 멤버를 랜덤하게 섞음
        for guild_name in self.group_dict.keys():
            self.guild_members[guild_name] = []  # 길드별 멤버 리스트 초기화

        for member in members:
            # 랜덤하게 길드 선택
            selected_guild = random.choice(list(self.group_dict.keys()))
            # 멤버를 선택한 길드에 추가
            self.group_dict[selected_guild].append(member)
            self.guild_members[selected_guild].append(member)
            print(f"{member}님이 {selected_guild} 길드에 추가되었습니다.")

    # 길드 정보 출력
    def display_guild_info(self):
        for guild_name, guild_members in self.guild_members.items():
            print(f"{guild_name} 길드: {', '.join(guild_members)}")

    # 길드별 점수 입력 및 차감 메서드
    def guild_score_method(self, guild_name, points):
        if guild_name in self.guild_score:
            self.guild_score[guild_name] += points
            print(f"{guild_name} 길드의 점수가 {points}점 {points > 0 and '증가' or '차감'}되었습니다.")
            # 길드에 속한 모든 멤버에 대해 퀘스트 점수 총점과 퀘스트 입력 총점에 점수를 추가
            print(guild_name)
            print(self.guild_members)
            for member in self.guild_members[guild_name]:
                for core_member in core_list:
                    if core_member['이름'] == member:
                        core_member["퀘스트 입력 총점"] += points
                        core_member["퀘스트 점수 총점"] += points
        else:
            print(f"{guild_name} 길드가 존재하지 않습니다.")



core_6 = Aiffel("민찬", 2023, "6기", "코어")
core_6.data_input()
core_6.get_id_num()
core_6.score("민찬", 2)
core_6.penalty("김건", -2)

core_6.remove_member("김경민")
core_6.remove_member("김영신")

core_6.display_info()

# Aiffel_Group 인스턴스 생성
group_manager = Aiffel_Group("Group Manager", 2023, "6기", "코어")

# 그룹핑 실행
group_manager.group_members()

# 그룹 정보 출력
group_manager.display_group_info()


# Aiffel_Guild 인스턴스 생성
guild_manager = Aiffel_Guild("Guild Manager", 2023, "6기", "코어")

# 길드 그루핑 실행
guild_manager.group_members()

# 길드 정보 출력
guild_manager.display_guild_info()

# 길드별 점수 조작
guild_manager.guild_score_method("힌튼", 5)
guild_manager.guild_score_method("알트만", -3)

# 전체 길드 점수 출력
print("전체 길드 점수:", guild_manager.guild_score)

core_6.display_info()
```
