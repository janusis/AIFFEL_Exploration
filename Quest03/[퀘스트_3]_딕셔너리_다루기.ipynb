```python
import re
from collections import defaultdict

# 사용자로부터 텍스트 파일명 입력 받기
print("="*3)
txt_file_name = input("입력값: \n")
file_url = '/content/drive/MyDrive/Aiffel/quest/Q03/'
file_path = file_url + txt_file_name
print("\n출력값 : ")

try:
    # 파일을 읽어서 텍스트 데이터를 가져옴
    with open(file_path, "r", encoding="utf-8") as file:
        text = file.read()
except FileNotFoundError:
    print("입력한 파일이 존재하지 않습니다.")
    exit()


# 모든 문자를 소문자로 변환
text = text.lower()

# 정규표현식을 사용하여 기호 제거 및 공백을 기준으로 단어로 분할
words = re.findall(r'\b\w+\b', text)

# 2-gram 페어 생성
ngrams = [(words[i], words[i+1]) for i in range(len(words)-1)]

# 2-gram 빈도를 저장할 딕셔너리 생성
ngram_freq = defaultdict(int)

# 빈도 계산
for ngram in ngrams:
    ngram_freq[ngram] += 1

# 가장 빈도가 높은 2-gram 페어 찾기
most_common_ngram = max(ngram_freq, key=ngram_freq.get)

# 딕셔너리를 빈도가 높은 순서로 정렬
sorted_ngram_freq = dict(sorted(ngram_freq.items(), key=lambda x: x[1], reverse=True))

print("max2gram:", most_common_ngram, ngram_freq[most_common_ngram])
print("dic :", sorted_ngram_freq)


```
